## データの読み込み
性別による単純な生存結果の予測を行う。<br>

```python
file_path = 'data/titanic_train.csv'
df = pd.read_csv(file_path)
df.head()
```

## データセットの概観

１２個の特徴量で、うち7個は数値データ。

## 性別を参照した単純な分類による生死の予測

男性であれば死（prediction=0）、女性であれば生（prediction=1）という単純な分類を行う。<br>
ディクショナリ形式の変数predictionsに、乗客の生死の予測結果を収納する。

## iterrowsの使い方  
pandas.DataFrame.iterrowsはタプルを返し、1番目の戻り値にnumerator(index),  
２番目の戻り値はDataFrameを返す。  
http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.iterrows.html

```python
predictions = {}
for passenger_index, passenger in df.iterrows():
    if passenger['Sex'] == 'male':
        predictions[passenger_index] = 0
    else:
        predictions[passenger_index] = 1
predictions
```

```python
# predictions[i]とdf['Survived']が合致するときは正解、合致しないときは不正解として、
# 正答率(correct_rate)を算出して、表示する。
total_count = 0
correct_count = 0
for row_index, row in df.iterrows():
    total_count += 1
    if predictions[row_index] == row['Survived']:
        correct_count += 1
correct_rate = correct_count/total_count
print(correct_rate)
```

## 結果

性別による単純な分類であっても、78%以上の精度で生存結果の予測が当たる。<br>
つまりタイタニック号事件では、女性を優先して救助したため、男性の生存率が低く、<br>
女性の生存率が高くなったことがわかる。  <br>

## 2. complex_heuristic.pyの解読

## もう少し複雑な分類
性別（Sex）、客室クラス（Pclass）、年齢（Age）、運賃（Fare）の違いで生死の予測を行う。  
以下の条件を満たす乗客を生（Prediction=1）とする。
1. 性別：女性（female）　または
2. 客室クラス：2級以下　かつ
3. 年齢：１３歳以下 かつ
4. 運賃：１０以上

```python
predictions = {}
# df = pandas.read_csv(file_path)
for passenger_index, passenger in df.iterrows():
    if ((passenger['Sex']=='female') or (passenger['Pclass']<=2) and
        (passenger['Age']<=13) and (passenger['Fare']>=10)):
        predictions[passenger_index] = 1
    else:
        predictions[passenger_index] = 0
```

```python
total_count = 0
correct_count = 0
for row_index, row in df.iterrows():
    total_count += 1
#     print(row_index, row['Survived'])
    if predictions[row_index] == row['Survived']:
        correct_count += 1
correct_rate = correct_count/total_count
print(correct_rate)
```

## 結果
上記のとおり、4つの条件を組み合わせると、80%以上の精度に改善することがわかる。  
ただしこの予測では、男性の生存者を抽出することができないので、これ以上の精度の  
改善は難しい。
